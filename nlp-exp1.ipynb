### **Task 02 ||**

```python
nltk.download('gutenberg')
from nltk.corpus import gutenberg
novel= gutenberg.raw('austen-persuasion.txt')
print("Selected novel:", 'austen-persuasion.txt')
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


words = word_tokenize(novel.lower())

stop_words = stopwords.words('english')
filtered_words = []
for word in words:
    if word not in stop_words and word.isalpha():
        filtered_words.append(word)

print("First 10 filtered words:", filtered_words[:10])
```

```python
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt

fdist = FreqDist(filtered_words)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(fdist)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### **Task 03 ||**

```python
nltk.download('gutenberg')

from nltk.corpus import gutenberg

#novel
novel = gutenberg.raw('carroll-alice.txt')
print("Selected novel:", 'carroll-alice.txt')
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Tokenize
words = word_tokenize(novel.lower())

# Remove stopwords then stemming
stop_words = stopwords.words('english')
ps = PorterStemmer()
filtered_stemmed_words = []
for word in words:
    if word not in stop_words and word.isalpha():
        stemmed_word = ps.stem(word)
        filtered_stemmed_words.append(stemmed_word)

print("First 10 filtered and stemmed words:", filtered_stemmed_words[:10])
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Before preprocessing
fdist_before = FreqDist(words)
top_15_before = fdist_before.most_common(15)
print("Top 15 words before preprocessing:")
for word, freq in top_15_before:
    print(f"{word}: {freq}")
```

```python
# Before
words_before, freqs_before = zip(*top_15_before)
plt.figure(figsize=(12, 6))
plt.bar(words_before, freqs_before)
plt.title("Top 15 Words Before Preprocessing")
plt.xticks(rotation=45)
plt.show()
```

```python
# After preprocessing (using filtered and stemmed words)
fdist_after = FreqDist(filtered_stemmed_words)
top_15_after = fdist_after.most_common(15)
print("\nTop 15 words after preprocessing:")
for word, freq in top_15_after:
    print(f"{word}: {freq}")
```

```python
words_after, freqs_after = zip(*top_15_after)
plt.figure(figsize=(12, 6))
plt.bar(words_after, freqs_after)
plt.title("Top 15 Words After Preprocessing")
plt.xticks(rotation=45)
plt.show()
```

### **Task 05 ||**

```python
nltk.download('udhr')
from nltk.corpus import udhr
hungarian_text = udhr.raw('Hungarian_Magyar-UTF8')
print("Selected text:", 'Hungarian_Magyar-UTF8')
```

```python
words = word_tokenize(hungarian_text.lower())
vowels = 'aeiouáéíóöőúüű'
vowel_sequences = []

for word in words:
    word_vowels = [char for char in word if char in vowels]
    if word_vowels:
        vowel_sequences.extend(word_vowels)

print("First 10 vowels extracted:", vowel_sequences[:10])
```

```python
from nltk import bigrams
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

vowel_bigrams = list(bigrams(vowel_sequences))
bigram_freq = FreqDist(vowel_bigrams)

print("Vowel bigram frequency distribution:")
for bigram, freq in bigram_freq.most_common():
    print(f"{bigram[0]}-{bigram[1]}: {freq}")
```

```python
top_bigrams = bigram_freq.most_common(15)
bigram_labels = [f"{bigram[0]}-{bigram[1]}" for bigram, _ in top_bigrams]
frequencies = [freq for _, freq in top_bigrams]

plt.figure(figsize=(12, 6))
plt.bar(bigram_labels, frequencies)
plt.title("Top 15 Vowel Bigram Frequencies")
plt.xlabel("Vowel Bigrams")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
